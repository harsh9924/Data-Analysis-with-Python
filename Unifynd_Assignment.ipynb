# **Question 6**

 Create a cohort of users with the following conditions and answer the following questions:
(Scan Count > 4 && Coupon Transaction Count > 2)
1. What’s the tier distribution of these users?
2. What’s the gender distribution of these users?
3. Which 5 stores have these users scanned the most from?
4. Which source has the most users coming from in this cohort?
5. Observe the behaviour of this cohort and write your observations and insights.

```python
from pandasql import sqldf
pysqldf = lambda q: sqldf(q, globals())
ques6 = pysqldf("SELECT tierId, customerScans.customerId, scanlogId, CouponId, gender, storeName  FROM couponTransactions, customerUser, customerScans WHERE couponTransactions.customerId = customerScans.customerId and customerUser.customerId = customerScans.customerId group by customerScans.customerId, customerScans.storeName")
print(ques6['tierId'].value_counts())
print(ques6['gender'].value_counts())
print(ques6['storeName'].head())
```

```python
df = pd.ques6(ques6.groupby(["Customer_ID"]).agg({'scanlogId':max, 'Sales_Amount':sum})).reset_index()
df.head()
```

**Solving by Cohort Method**
- Clean the data.
- put transactions performed by the same customer on the same date in a single row since our demonstration will be based on transaction date, quantity, and sales amount.
- To run a cohort analysis, we’ll need to: 
  Split the data into groups that can be analyzed on the basis of time
  
  Assign a cohort index for each transaction
  
  Create two new columns
- finding out how many unique customers were acquired per month per cohort by setting the cohort_idx per row. 
- heatmap functions provided by matplotlib, we can see a graphical representation of the number of unique customers per cohort over time.
- Execute Retention Analysis.

**Observations & Insights**
- Tier 2 group has the most number of transactions. 
- As we can see the gender distribution Male transactions are 50% more than the female transactions.
- From the data we can see that Store "Annavillas" is the place where most scanning is done.
- It shows most of the coupons were scanned at Food place which shows people tend to spend more on food than other commodities.
 

# **Question 7**

Month on Month comparison:

```python
from pandasql import sqldf
pysqldf = lambda q: sqldf(q, globals())
#print(pysqldf("SELECT tierId, COUNT(DISTINCT customerUser.customerId) as UniqueUser, COUNT(couponId) as TotalCouponUnlocked, AVG(earnedPoints)as AverageEarnedPoints, AVG(burnedPoints)as AverageBurnedPoints FROM couponTransactions INNER JOIN customerUser on couponTransactions.customerId = customerUser.customerId group by tierId")) 
ques5 = pysqldf ('''Select * From (SELECT tierId, COUNT(DISTINCT customerUser.customerId)as sept_uniqueUser, COUNT(customerScans.scanlogId)as sept_billScan, SUM(customerScans.billTotal)as sept_scanTotal, COUNT(couponTransactions.couponId)as sept_unlocks FROM customerScans, couponTransactions, customerUser WHERE customerScans.customerId = couponTransactions.customerId and couponTransactions.customerId = customerUser.customerId and billDate between'2020-09-01' and '2020-10-01' and couponUnlockDate between'2020-09-01' and '2020-10-01' group by tierId) as Sept_Table, 
              (SELECT tierId, COUNT(DISTINCT customerUser.customerId)as oct_uniqueUser, COUNT(customerScans.scanlogId)as oct_billscan, SUM(customerScans.billTotal)as oct_scanTotal, COUNT(couponTransactions.couponId)as oct_unlocks FROM customerScans, couponTransactions, customerUser WHERE customerScans.customerId = couponTransactions.customerId and couponTransactions.customerId = customerUser.customerId and billDate between'2020-09-01' and '2020-10-01' and couponUnlockDate between'2020-10-01' and '2020-11-01' group by tierId) as Oct_Table where Sept_Table.tierId = Oct_Table.tierId''')
print(ques5)
print(ques5.sum(axis = 0))
```
